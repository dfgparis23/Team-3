```python
won_opportunities = relevant_col_qs_1.where(relevant_col_qs_1["Stage"] == "Won")
```

```python
cities_donations = won_opportunities[["Amount received", "Postal Code"]]
sectors_donations = won_opportunities[["Amount received", "Activity sector"]]
business_accounts = won_opportunities[["Amount received", "Business Account"]]
```

```python
sum_cities_amount = cities_donations.groupby("Postal Code")["Amount received"].sum().abs().reset_index()
```

```python
sum_cities_amount
```

```python
sum_sectors_amount = sectors_donations.groupby("Activity sector")["Amount received"].sum().abs().reset_index()
```

```python
sum_ba_amount = business_accounts.groupby("Business Account")["Amount received"].sum().abs().reset_index()
```

```python
mean_diff_cities = sum_cities_amount["Amount received"].median()
mean_diff_sectors = sum_sectors_amount["Amount received"].median()
mean_diff_ba = sum_ba_amount["Amount received"].median()
```

```python
import matplotlib.pyplot as plt


def plot_data(data_frame, title, col, mean_val):
    plt.figure(figsize=(8, 6))
    plt.bar(data_frame[col], data_frame["Amount received"])
    plt.axhline(y=mean_val, color='red', linestyle='--', label=f'Mean: {mean_val}')
    plt.xlabel(f"Global amount given by each {col}")
    plt.ylabel("Amount received")
    plt.title(title)
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Plot for cities_donations
plot_data(sum_cities_amount, "departments_donations", "Postal Code", mean_diff_cities)

# Plot for sectors_donations
plot_data(sum_sectors_amount, "sectors_donations", "Activity sector", mean_diff_sectors)

# Plot for business_accounts
plot_data(sum_ba_amount, "business_accounts", "Business Account", mean_diff_ba)
```

```python
relevant_col_qs_1["Year"] = relevant_col_qs_1["Actual Close Date"].dt.year
```

```python
won_opportunities["Year"] = won_opportunities["Actual Close Date"].dt.year
```

```python
cities_opp_count = relevant_col_qs_1.groupby(["Postal Code","Year"]).size().reset_index(name="Count")
cities_won_count = won_opportunities.groupby(["Postal Code", "Year"]).size().reset_index(name="Count")
```

```python
ba_opp_count = relevant_col_qs_1.groupby(["Business Account", "Year"]).size().reset_index(name="Count")
ba_won_count = won_opportunities.groupby(["Business Account", "Year"]).size().reset_index(name="Count")
```

```python
sectors_opp_count = relevant_col_qs_1.groupby(["Activity sector", "Year"]).size().reset_index(name="Count")
sectors_won_count = won_opportunities.groupby(["Activity sector", "Year"]).size().reset_index(name="Count")
```

```python
cities_opp_count_2021 = cities_opp_count.where(cities_opp_count["Year"] == 2021).dropna()
cities_opp_count_2022 = cities_opp_count.where(cities_opp_count["Year"] == 2022).dropna()
cities_won_count_2021 = cities_won_count.where(cities_won_count["Year"] == 2021).dropna()
cities_won_count_2022 = cities_won_count.where(cities_won_count["Year"] == 2022).dropna()
```

```python
sectors_opp_count_2021 = sectors_opp_count.where(sectors_opp_count["Year"] == 2021).dropna()
sectors_opp_count_2022 = sectors_opp_count.where(sectors_opp_count["Year"] == 2022).dropna()
sectors_won_count_2021 = sectors_won_count.where(sectors_won_count["Year"] == 2021).dropna()
sectors_won_count_2022 = sectors_won_count.where(sectors_won_count["Year"] == 2022).dropna()
```

```python
ba_opp_count_2021 = ba_opp_count.where(sectors_opp_count["Year"] == 2021).dropna()
ba_opp_count_2022 = ba_opp_count.where(sectors_opp_count["Year"] == 2022).dropna()
ba_won_count_2021 = ba_won_count.where(sectors_won_count["Year"] == 2021).dropna()
ba_won_count_2022 = ba_won_count.where(sectors_won_count["Year"] == 2022).dropna()
```

```python
merged_cities_2021 = cities_opp_count_2021.merge(cities_won_count_2021, on="Postal Code", suffixes=('_opp', '_won'))
merged_cities_2021["Quotient"] = merged_cities_2021["Count_won"] / merged_cities_2021["Count_opp"]
print(merged_cities_2021)
```

```python
merged_cities_2022 = cities_opp_count_2022.merge(cities_won_count_2022, on="Postal Code", suffixes=('_opp', '_won'))
merged_cities_2022["Quotient"] = merged_cities_2022["Count_won"] / merged_cities_2022["Count_opp"]
print(merged_cities_2022)
```

```python
merged_sectors_2021 = sectors_opp_count_2021.merge(sectors_won_count_2021, on="Activity sector", suffixes=('_opp', '_won'))
merged_sectors_2021["Quotient"] = merged_sectors_2021["Count_won"] / merged_sectors_2021["Count_opp"]
print(merged_sectors_2021)
```

```python
merged_sectors_2022 = sectors_opp_count_2022.merge(sectors_won_count_2022, on="Activity sector", suffixes=('_opp', '_won'))
merged_sectors_2022["Quotient"] = merged_sectors_2022["Count_won"] / merged_sectors_2022["Count_opp"]
print(merged_sectors_2022)
```

```python
mean_cities_2021 = merged_cities_2021["Quotient"].mean()
mean_cities_2022 = merged_cities_2022["Quotient"].mean()
mean_sectors_2021 = merged_sectors_2021["Quotient"].mean()
mean_sectors_2022 = merged_sectors_2022["Quotient"].mean()
```

```python
import matplotlib.pyplot as plt


def plot_quotient(data_frame, col, title, mean_val):
    data_frame = data_frame.sort_values(by="Quotient", ascending=True)
    plt.figure(figsize=(8, 6))
    plt.bar(data_frame[col], data_frame["Quotient"])
    plt.axhline(y=mean_val, color='red', linestyle='--', label=f'Mean: {mean_val}')
    plt.xlabel(col)
    plt.ylabel("Quotient of won opportunities")
    plt.title(title)
    plt.xticks(rotation=90)
    plt.show()
```

```python
plot_quotient(merged_cities_2021, "Postal Code", "Quotient of won opportunities by the year 2021 by department", mean_cities_2021)
plot_quotient(merged_cities_2022, "Postal Code", "Quotient of won opportunities by the year 2022 by department", mean_cities_2022)
plot_quotient(merged_sectors_2021, "Activity sector", "Quotient of won opportunities by the year 2021 by sector", mean_sectors_2021)
plot_quotient(merged_sectors_2022, "Activity sector", "Quotient of won opportunities by the year 2022 by sector", mean_sectors_2022)
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

geojson_url = "https://france-geojson.gregoiredavid.fr/repo/departements.geojson"

gdf = gpd.read_file(geojson_url)

highlighted_departments = ["07", "11", "17", "21", "22", "35", "61", "89"]
highlighted_gp_departments = ["02", "04", "13", "14", "18", "38", "73", "75", "77", "84"]

highlighted_gdf = gdf[gdf['code'].isin(highlighted_departments)]
highlighted_gp_gdf = gdf[gdf['code'].isin(highlighted_gp_departments)]

ax = gdf.plot(figsize=(10, 10), edgecolor='k', facecolor='none')
plt.title("Map of France with Departments")

highlighted_gdf.plot(ax=ax, color='red', alpha=0.5, label='Highlighted Departments')
highlighted_gp_gdf.plot(ax=ax, color='green', alpha=0.5, label='Highlighted GP Departments')

plt.legend()
plt.axis('off')
plt.show()
```
