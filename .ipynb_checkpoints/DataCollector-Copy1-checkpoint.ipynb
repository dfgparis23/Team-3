```python
relevant_col_qs_1 = question1_df[["Business Account", "Stage", "Amount received", "Created On_x", "Actual Close Date", "Postal Code", "Country Name", "Activity sector" ]]
```

```python
relevant_col_qs_1.isna().sum()
```

```python
relevant_col_qs_1["Amount received"].fillna(0, inplace = True)
```

```python
replacement_date = pd.to_datetime("2023-10-20")
relevant_col_qs_1["Actual Close Date"].fillna(replacement_date, inplace = True)
```

```python
relevant_col_qs_1.isna().sum()
```

```python
relevant_col_qs_1["Postal Code"].unique()
```

```python
relevant_col_qs_1.dropna(subset=["Postal Code"], inplace = True)
```

```python
relevant_col_qs_1.isna().sum()
```

```python
relevant_col_qs_1["Activity sector"].unique()
```

```python
relevant_col_qs_1["Activity sector"].fillna("NO RELATED TO ANY SECTOR", inplace = True)
```

```python
relevant_col_qs_1.isna().sum()
```

```python
won_opportunities = relevant_col_qs_1.where(relevant_col_qs_1["Stage"] == "Won")
```

```python
won_opportunities
```

```python
cities_donations = won_opportunities[["Amount received", "Postal Code"]]
sectors_donations = won_opportunities[["Amount received", "Activity sector"]]
business_accounts = won_opportunities[["Amount received", "Business Account"]]
```

```python
sum_cities_amount = cities_donations.groupby("Postal Code")["Amount received"].sum().abs().reset_index()
```

```python
sum_cities_amount
```

```python
sum_sectors_amount = sectors_donations.groupby("Activity sector")["Amount received"].sum().abs().reset_index()
```

```python
sum_ba_amount = business_accounts.groupby("Business Account")["Amount received"].sum().abs().reset_index()
```

```python
import matplotlib.pyplot as plt


def plot_difference(data_frame, title, col):
    plt.figure(figsize=(8, 6))
    plt.bar(data_frame[col], data_frame["Amount received"])
    plt.xlabel(f"Global amount given by each {col}")
    plt.ylabel("Amount received")
    plt.title(title)
    plt.xticks(rotation=90)
    plt.show()
```

```python
# Plot for cities_donations
plot_difference(sum_cities_amount, "cities_donations", "Postal Code")

# Plot for sectors_donations
plot_difference(sum_sectors_amount, "sectors_donations", "Activity sector")

# Plot for business_accounts
plot_difference(sum_ba_amount, "business_accounts", "Business Account")
```
