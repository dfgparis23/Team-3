```python
import pandas as pd
import numpy as np
```

```python
business_account = pd.read_excel("../shared_data_read_only/Business Account.xlsx")
consumer_price_index = pd.read_excel("../shared_data_read_only/consumer_price_index.xlsx")
contacts = pd.read_excel("../shared_data_read_only/Contacts.xlsx")
data_dictionary = pd.read_excel("../shared_data_read_only/Data dictionary.xlsx")
invoice_opportunities = pd.read_excel("../shared_data_read_only/Invoice and Opportunities.xlsx")
items_opportunities = pd.read_excel("../shared_data_read_only/Items and Opportunities.xlsx")
non_stock_items = pd.read_excel("../shared_data_read_only/Non-Stock Items.xlsx")
opportunities = pd.read_excel("../shared_data_read_only/Opportunities.xlsx")
```

```python
business_data = business_account[business_account['Type'] == 'Business Account']
business_data_clean = business_account[business_account['Customer Status'] != 'Inactive']
business_data_clean.info()
```

```python
df_merged_invoice = pd.merge(opportunities, invoice_opportunities, on='Opportunity ID', how='inner')
big_data = pd.merge(df_merged_invoice, business_data_clean, on='Business Account', how='inner')
data_analysed = big_data[big_data['Status'] == 'Won']
data_received = data_analysed.groupby("Business Account")["Amount received"].sum()
data_estimated = data_analysed.groupby("Business Account")["Estimated amount"].sum()
prop = (data_received / data_estimated).round(2)
prop = prop[prop <= 1]
filtered_data = prop[prop < 0.6]
filtered_data 
```

```python
prop.plot(kind='line', figsize=(12, 6), marker='o', color='skyblue')
plt.title('Proportion of Amount Received to Estimated Amount Over Time')
plt.xlabel('BA')
plt.ylabel('Proportion')
plt.axhline(y=0.6, color='red', linestyle='--', label='Threshold (0.6)')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
mean_prop = np.mean(prop)
var_prop = np.var(prop, ddof=1)  # ddof=1 for sample variance

alpha_initial = mean_prop * ((mean_prop * (1 - mean_prop)) / var_prop - 1)
beta_initial = alpha_initial * (1 / mean_prop - 1)

def log_likelihood(params, data):
    alpha, beta = params
    log_likelihood = np.sum(np.log(np.power(data, alpha - 1) * np.power(1 - data, beta - 1)))
    print("Params:", params, "Log-Likelihood:", log_likelihood)
    return -log_likelihood

result = minimize(log_likelihood, [alpha_initial, beta_initial], args=(prop,), method='L-BFGS-B')
alpha_estimated, beta_estimated = result.x
print(f"Estimated alpha: {alpha_estimated}")
print(f"Estimated beta: {beta_estimated}")
x = np.linspace(0, 1, 100)
pdf = np.power(x, alpha_estimated - 1) * np.power(1 - x, beta_estimated - 1)
pdf /= np.sum(pdf)  
plt.plot(x, pdf, label='Estimated Beta Distribution')
plt.hist(filtered_data, bins=20, density=True, alpha=0.5, label='Filtered Data Histogram')
plt.legend()
plt.title('Estimated Beta Distribution Parameters')
plt.xlabel('Proportion')
plt.ylabel('Probability Density')
plt.show()
```
